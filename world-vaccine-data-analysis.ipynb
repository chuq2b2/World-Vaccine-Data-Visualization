```python
import numpy as np
import pandas as pd

df_vaccination_status = pd.read_csv("Worldwide Vaccine Data.csv")

df_vaccination_status
```

```python
countries = df_vaccination_status["Country"]
vaccinated_status = df_vaccination_status["% of population vaccinated"]
fully_vaccinated_status = df_vaccination_status["% of population fully vaccinated"]
```

```python
import matplotlib.pyplot as plt 

#Bar chart for vaccinated status 
plt.figure(figsize=(20, 10))
plt.bar(countries, vaccinated_status, color='skyblue', alpha=0.75)
plt.xlabel('Country')
plt.ylabel('% of Population Vaccinated')
plt.title('Vaccination Status by Country')
plt.xticks(rotation=90) 
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100%
plt.tight_layout()
plt.show()
```

```python
#Bar chart for fully vaccination status 
plt.figure(figsize=(20, 10))
plt.bar(countries, fully_vaccinated_status, color='blue', alpha=0.75)
plt.xlabel('Country')
plt.ylabel('% of Population Vaccinated')
plt.title('Fully Vaccination Status by Country')
plt.xticks(rotation=90) 
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100%
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Assuming df_vaccination_status is your DataFrame containing country-wise vaccination data

# Create an array of indices for the countries
x = np.arange(len(countries))

# Width of each bar
bar_width = 0.35

plt.figure(figsize=(20, 10))

# Bar chart for overall vaccinated status
plt.bar(x - bar_width/2, vaccinated_status, bar_width, color='skyblue', alpha=0.75, label='Vaccinated')
# Bar chart for fully vaccinated status
plt.bar(x + bar_width/2, fully_vaccinated_status, bar_width, color='blue', alpha=0.75, label='Fully Vaccinated')

plt.xlabel('Country')
plt.ylabel('% of Population Vaccinated')
plt.title('Vaccination Status by Country')
plt.xticks(x, countries, rotation=90)
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid lines
plt.legend()
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100%
plt.tight_layout()
plt.show()
```

```python
#Number of Countries in Different Vaccination Status Ranges
below_25 = df_vaccination_status[df_vaccination_status['% of population vaccinated'] < 25]['Country'].count()
between_25_50 = df_vaccination_status[(df_vaccination_status['% of population vaccinated'] >= 25) & 
                                      (df_vaccination_status['% of population vaccinated'] < 50)]['Country'].count()
between_50_75 = df_vaccination_status[(df_vaccination_status['% of population vaccinated'] >= 50) & 
                                      (df_vaccination_status['% of population vaccinated'] < 75)]['Country'].count()
above_75 = df_vaccination_status[df_vaccination_status['% of population vaccinated'] >= 75]['Country'].count()

labels = ['Below 25%', 'Between 25% and 50%', 'Between 50% and 75%', 'Above 75%']
counts = [below_25, between_25_50, between_50_75, above_75]

plt.figure(figsize=(8, 6))
plt.grid(True, linestyle='--', alpha=0.3)  

bars = plt.bar(labels, counts, color=['red', 'skyblue', 'lightgreen', 'green'])

for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), count,
             ha='center', va='bottom', fontsize=10)

plt.bar(labels, counts, color=['red', 'skyblue', 'lightgreen', 'green'])
plt.xlabel('Vaccination Status')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.xticks(range(len(labels)), labels)
plt.title('Number of Countries in Different Vaccination Status Ranges')
plt.show()
```
